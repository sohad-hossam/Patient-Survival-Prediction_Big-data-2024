```python
# Create a SparkSession
spark = SparkSession.builder \
    .appName("PySpark-Script") \
    .getOrCreate()
```

### Preprocessing the data

```python
dataset = spark.read.csv("dataset.csv", header=True, inferSchema=True)
dataset = dataset.drop("_c83")
dataset.printSchema()
print(dataset.count())
# print(dataset.rdd)
```

```python
'''
Data Cleaning:
1) Removing missing data/ N/A / unexpected values --> Done
2) dealing with outliers --> Done
3) data normalization/standardization -->  
4) transform features to categorial if needed --> Done
5) visualizing the data using plots(box/scatter/histograms ...etc) --> Done
'''
dataset_without_na = dataset.na.drop(how='any')
print(dataset_without_na.count())
 
# ---------------------- For illustration reasons ----------------------#
# col_names = [name for name, dtype in dataset_without_na.dtypes if dtype == "string"]
# for col_name in col_names:
#     dataset_without_na.groupby(col_name).count().show(truncate=False)

col_names = [name for name, dtype in dataset_without_na.dtypes if dtype == "string"]
for col_name in col_names:
    col_unique_values = dataset_without_na.select(col_name).distinct()
    if (col_unique_values.count() > 2 ) : 
        for col_value in col_unique_values.toLocalIterator():
            dataset_without_na = dataset_without_na.withColumn(col_value[0],(dataset_without_na[col_name]==col_value[0]).cast("int"))
        dataset_without_na = dataset_without_na.drop(col_name) 
    else:
        col_unique_values = dataset_without_na.select(col_name).distinct()
        dataset_without_na = dataset_without_na.withColumn(col_name, (dataset_without_na[col_name]=='F').cast("int"))

dataset_without_na.show()
dataset_converted = dataset_without_na.toPandas()
# dataset_without_na.show()
```

```python
for col_name, dtypes in dataset_converted.dtypes.items():
    if dtypes != 'object' and len(dataset_converted[col_name].unique()) > 2:
        plt.xlabel("X-axis")
        plt.ylabel("Y-axis")
        plt.title(col_name)
        plt.boxplot(dataset_converted[col_name])
        plt.show()
        plt.hist(dataset_converted[col_name], bins=100)
        plt.show()

# ay 7aga id, apache_4a_hospital_death_prob, apache_4a_hospital_death_prob_icu
```

```python
bmi_hospital_death_count = dataset_converted[['bmi', 'hospital_death', 'patient_id']].groupby(['bmi', 'hospital_death']).count()
print(bmi_hospital_death_count)

min_height = dataset_converted['height'].min()
print(min_height)

bmi_hospital_death_count = dataset_converted[['weight', 'hospital_death', 'patient_id']].groupby(['weight', 'hospital_death']).count()
print(bmi_hospital_death_count)

apache_4a_hospital_death_prob_min = dataset_converted['apache_4a_hospital_death_prob'].max()
print(apache_4a_hospital_death_prob_min)
```

```python
num_columns = len(dataset_without_na.columns)
print("Number of columns:", num_columns)

dataset_without_outliers = dataset_without_na.filter(dataset_without_na['apache_4a_icu_death_prob'] >= 0)
dataset_without_outliers = dataset_without_outliers.filter(dataset_without_outliers['apache_4a_hospital_death_prob'] >= 0)
dataset_without_outliers = dataset_without_outliers.drop(*['encounter_id', 'patient_id', 'hospital_id', 'icu_id']).toPandas()

num_columns = len(dataset_without_outliers.columns)
print("Number of columns:", num_columns)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title('apache_4a_icu_death_prob')
plt.boxplot(dataset_without_outliers['apache_4a_icu_death_prob'])
plt.show()
plt.hist(dataset_without_outliers['apache_4a_icu_death_prob'], bins=100)
plt.show()

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title('apache_4a_hospital_death_prob')
plt.boxplot(dataset_without_outliers['apache_4a_hospital_death_prob'])
plt.show()
plt.hist(dataset_without_outliers['apache_4a_hospital_death_prob'], bins=100)
plt.show()

```

```python
dataset_without_outliers_dataframe = spark.createDataFrame(dataset_without_outliers)

# checking if duplicates exist
# count_before = dataset_without_outliers.count()
# df_no_duplicates = dataset_without_outliers.dropDuplicates().count()
# print(count_before, df_no_duplicates)

# for column in dataset_without_outliers_dataframe.columns:
#     if dataset_without_outliers_dataframe.filter(dataset_without_outliers_dataframe[column] == float('inf')).count() > 0:
#         print(f"There are infinity values in column '{column}'")
#     else:
#         print(f"No infinity values in column '{column}'")
```

```python
assembler = VectorAssembler(inputCols=dataset_without_outliers_dataframe.columns, outputCol="features")
dataset_scaled = assembler.transform(dataset_without_outliers_dataframe)
dataset_scaled.show()

for column in dataset_scaled.columns:
    standardizer = StandardScaler(inputCol=column, outputCol=column, withStd=True, withMean=True)
    dataset_scaled = standardizer.fit(dataset_scaled).transform(dataset_scaled)
dataset_scaled.show()
```

```python
for column in dataset_without_outliers_dataframe.columns:
    assembler = VectorAssembler(inputCols=[column], outputCol=column+"_vec")
    dataset_scaled = assembler.transform(dataset_without_outliers_dataframe)
    standardizer = StandardScaler(inputCol=column+"_vec", outputCol=column+"_scaled", withStd=True, withMean=True)
    dataset_scaled = standardizer.fit(dataset_scaled).transform(dataset_scaled)
dataset_scaled.show()
```
